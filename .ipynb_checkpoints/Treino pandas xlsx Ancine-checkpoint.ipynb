# Analide de dados de Filmes exibidos no brasil de 2009 a 2017
Com base nos dados da Ancine - http://oca.ancine.gov.br 

Arquivo -> 2120_0.xlsx

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel('2120_0.xlsx',skiprows=2,skipfooter=16)
df.head(4)
```

```python
df.info()
```

```python
df.tail(2)
```

```python
# DataFrame com os filmes exibidos em 2009
df_09 = df[df['Ano de exibição'] == 2009]
```

```python
# DataFrame com os filmes Brasileiros exibidos em 2009
df_09_br = df_09[df_09['Nacionalidade da obra'] == 'Brasileiros']
```

```python
# Quais os 10 filmes brasileiros tiveram a melhor bilheteria no ano de 2009?
df_09_br.nlargest(10, 'Renda (R$) no ano de exibição')
```

```python
# Qual o valor recebido por exibiçao (renda/público)
df_09_br['Valor exibição'] = df_09_br['Renda (R$) no ano de exibição'] / df_09_br['Público no ano de exibição']
```

```python
df_09_br.head(2)
```

```python
# Quais os 10 filmes que receberam mais por exibição?
df_09_br.nlargest(10, 'Valor exibição')
```

```python
# Quais os 10 filmes que receberam menos por exibição?
df_09_br.nsmallest(10, 'Valor exibição')
```

```python
# Quem distribuiu mais filmes em 2009?
df_09_br['Distribuidora'].value_counts().head(1)
```

```python
# Qual o filme brasileiro com maior público em 2009?
df_09_br.nlargest(1, 'Público no ano de exibição')
```

```python
# Qual o maior valor recebido por uma distribuidora em 2009?
df_09_br.groupby(by='Distribuidora', sort=True,)['Renda (R$) no ano de exibição'].max()
```

```python
df_09_br['Distribuidora'].value_counts().count()
```

```python
# Quanto cada distribuidora recebeu no total, pelos seus filmes brasileiros distribuidos no ano de 2009?
df_09_br.groupby(by='Distribuidora', sort=True,)['Renda (R$) no ano de exibição'].sum()
```

```python
# Filmes brasileiros de 2009, Warner
df_09_br[df_09_br['Distribuidora'] == 'Warner']
```

```python
# Quais as 5 distribuidoras que mais receberam por distribuição de filmes brasileiros no ano de 2009 e quanto foi recebido por cada uma?
# Solução: Abrir o df original, criar um df apenas com o ano de distribuição, depois outro df apenas com filmes brasileiros
# Agrupar por 'Distribuidora', pegar o resultado da 'Renda (R$) no ano de exibição', usar a função sum() e depois a função nlargest()
df_09_br.groupby(by='Distribuidora')['Renda (R$) no ano de exibição'].sum().nlargest(5)
```

```python
df_09['Origem da empresa distribuidora'].value_counts()
```

```python
# df específico com dados de 2009, filmes brasileiros, distribuição nacional
df_09_br_nc = df_09_br[df_09_br['Origem da empresa distribuidora'] == 'Distribuição Nacional']
```

```python
df_09_br_nc.info()
```

```python
df_09_br_nc.head(3)
```

```python
# Qual a soma do valor recebido pelas 5 primeiras distribuidoras nacionais, exibindo filmes nacionais no ano de 2009?
df_09_br_nc.groupby(by='Distribuidora')['Renda (R$) no ano de exibição'].sum().nlargest(5)
```

```python
df_09_dt.head(3)
```

```python
# Distribuidoras nacionais no quadro da ANCINE
df_09_dt['Distribuidora'].value_counts().count()
```

```python
# 5 Distribuidoras nacionais que mais faturaram no ano de 2009?
df_09_dt.groupby(by='Distribuidora')['Renda (R$) no ano de exibição'].sum().nlargest(5)
```
