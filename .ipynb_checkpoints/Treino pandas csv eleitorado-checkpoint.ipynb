```python
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
# Quais municípios tem mais homens do que mulheres?
# Criar um df apenas com os municípios onde haja mais eleitores homens que mulheres
df_m = df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']]
```

```python
# Nova coluna com relacao entre eleitoras e eleitores
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# 5 municípios com maior quantidade de eleitoras em relação aos eleitores
df.nlargest(5, 'RELACAO_FM').T
```

```python
%matplotlib inline
# Plotar um gráfico de distribuição da relação f/m
df['RELACAO_FM'].plot.hist(bins=100)
```

```python
# Importar as bibliotecas de gráficos
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotar a distribuição da relação f/m
sns.distplot(df['RELACAO_FM'], bins=100, color='red')
```

```python
# Retirando a linha 'kde=False'
sns.distplot(df['RELACAO_FM'], bins=100, color='red', kde=False)

# Título para o gráfico
plt.title('Relação Eleitoras/Eleitores', fontsize=18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.axvline(1.0, color='blue', linestyle='--')
```

```python
# Verificar a distribuição masc/fem mostrando os pontos(municipios)
sns.swarmplot(data=df, x='NM_REGIAO', y='RELACAO_FM')
plt.axhline(1.0, color='blue', linestyle='-')
```

```python
sns.stripplot(data=df, x='NM_REGIAO', y='RELACAO_FM')
```

```python
# Vamos plotar a quantidade total de eleitores por faixa etária
# usando gráfico de barras horizontal
# passo1: listar as colunas de interesse
colunas = ['QTD_ELEITORES_16','QTD_ELEITORES_17','QTD_ELEITORES_18','QTD_ELEITORES_19','QTD_ELEITORES_20']

total_idade = df[colunas].sum()
```
