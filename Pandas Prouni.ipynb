```python
#Quantidade de cursos por UF
df['uf_busca'].value_counts()
```

```python
# Qual o curso com a mensalidade mais alta?
df[df['mensalidade'] > 9000.00 ]
```

```python
# Em quais Estados e cidades estão os cursos mais baratos do Brasil?
df[df['mensalidade'] == df['mensalidade'].min()][['uf_busca','cidade_busca']]
```

```python
# Qual a mensalidade mais cara por UF
# Agrupar por 'uf_busca', coluna ['mensalidade']
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Mostrar todos os dados dos cursos mais caros de cada UF
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()][['uf_busca', 'mensalidade']]
```

```python
# Mensalidades entre 500 e 1000
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)]
```

```python
# Quais as maiores notas por UF
df.loc[df.groupby('uf_busca')['nota_integral_ampla'].idxmax()][['uf_busca', 'turno', 'mensalidade', 'nota_integral_ampla']]
```

```python
# Qual a mensalidade mais cara, no período noturno, paga para cada Estado?
# Primeiro definir um df apenas do período noturno
df_not = df[df['turno'] == 'Noturno']
# Localizar por UF, a mensalidade mais cara e selecionar as colunas a serem exibidas
df_not.loc[df_not.groupby('uf_busca')['mensalidade'].idxmax()][['uf_busca', 'turno', 'mensalidade', 'nota_integral_ampla', 'nome', 'universidade_nome']]
```

```python
# Qual o valor mais caro para um curso de enfermagem? Qual a média de valor entre os cursos do Brasil?
# Criar um df específico para enfermagem.
df_enf = df[df['nome'] == 'Enfermagem']
df_enf.head(2)
```

```python
# Achar o valor mais caro
df_enf.max()[['uf_busca', 'nome', 'mensalidade']]
```

```python
# Valor médio das mensalidades por UF
df_enf.groupby('uf_busca').mean()['mensalidade']
```

```python
# Qual o valor pago nas mensalidades no Ceará?
df_enf.loc[df_enf['uf_busca'] == 'CE'][['turno','universidade_nome','mensalidade']]
```

```python
df[df['universidade_nome'] == 'Faculdade Princesa do Oeste - FPO']
```

```python
df_enf.nlargest(5,'mensalidade')
```
