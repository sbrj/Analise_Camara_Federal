```python
import pandas as pd
import requests
```

```python
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados?itens=100&ordem=ASC&ordenarPor=nome'
resposta = requests.get(url)
```

```python
# Criando um df com base no json
df = pd.DataFrame(resposta.json()['dados'])
```
